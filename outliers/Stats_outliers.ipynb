# Heart Disease Prediction. Outliers

### Import neccesary libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from scipy import stats
import numpy as np
```

### Read data from 'outliers.csv' file

```python
data = pd.read_csv("outliers.csv")
data.head(10)
```

### Age distribution

```python
sns.distplot(data['age'],rug=True)
plt.show()
```

### Sex Distribution

```python
sns.countplot(x='sex', data=data)
plt.xlabel("Sex (0 = female, 1= male)")
plt.show()
```

```python
pd.crosstab(data.sex,data.target).plot(kind="bar",figsize=(10,5))
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.xticks(rotation=0)
plt.legend(["Haven't Disease", "Have Disease"])
plt.ylabel('Frequency')
plt.show()
```

### Fasting blood sugar

Fasting blood sugar >120
- 0 - false
- 1 - true

```python
sns.countplot(x='fbs', data=data)
plt.xlabel('FBS - (Fasting Blood Sugar > 120 mg/dl)')
plt.show()
```

```python
pd.crosstab(data.fbs,data.target).plot(kind="bar",figsize=(10,5))
plt.xlabel('FBS - (Fasting Blood Sugar > 120 mg/dl) (1 = true; 0 = false)')
plt.xticks(rotation = 0)
plt.legend(["Haven't Disease", "Have Disease"])
plt.ylabel('Frequency of Disease or Not')
plt.show()
```
